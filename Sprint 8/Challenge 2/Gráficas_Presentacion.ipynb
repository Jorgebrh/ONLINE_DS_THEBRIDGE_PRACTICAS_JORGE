```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from bootcampviztools import *
sns.set(color_codes = True)
```

```python
df_titanic = pd.read_csv("./data/titanic.csv")
```

```python
#1.
supervivientes = df_titanic["survived"].value_counts(normalize = True) * 100
supervivientes.index= supervivientes.index.map({0: "Falleció", 1: "Sobrevivió"})
fig, ax= plt.subplots(1, 1, figsize = (10,5))
barras = sns.barplot(x = supervivientes.index, y = supervivientes.values,palette= "colorblind", ax=ax)
ax.set_xlabel("")
for container in barras.containers:
    ax.bar_label(container, padding=5, fmt="%.1f%%")
plt.show()
```

```python
#2
sns.catplot(data=df_titanic,
    x="sex",
    y="age",
    hue="alive",
    kind="box",
    height=6,
    aspect=1.2
);
```

```python
#3
ordenadas = ["First","Second","Third"]
g = sns.catplot(x= "alive", col = "class", hue="who", data= df_titanic, kind = "count", col_order= ordenadas)
plt.show()
```

```python
#4
ordenadas = ["First","Second","Third"]
g = sns.catplot(x="alive",  hue="who", data=df_titanic[df_titanic["who"].isin(["woman", "child"])],
    kind="count",
    col_order= ordenadas)

plt.show()
```

```python
#5
ordenadas = ["First","Second","Third"]
g = sns.catplot(data=df_titanic,
    x="class",
    hue="alive",
    col="embark_town",
    kind="count",
    order=ordenadas,
    col_order=["Cherbourg", "Queenstown", "Southampton"]
)
```
